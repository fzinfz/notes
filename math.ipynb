```python
#Current Transformer
import math
i_rms_current = 60
nb_turns = 2000
U_sensor = 1
I_measured = math.sqrt(2) * i_rms_current
I_sensor = I_measured / nb_turns 
print(I_sensor)
R_burden = U_sensor / I_sensor
print(R_burden)
P = I_sensor * I_sensor * R_burden
print(P)
```
